```python
from graphviz import Digraph,Graph

dot = Digraph(name='knights',comment='The Round Table',engine='neato')

dot  #doctest: +ELLIPSIS

dot.node('A', 'Arthur')
dot.node('B', 'Bedevere')
dot.node('C', 'Calliope')
dot.edge('A','B')
dot.edge('B','A')
dot.edge('C','A')
```

```python
print(dot.source)
```

```python
dot.render('test-output/round-table.gv', view=True) 
```
